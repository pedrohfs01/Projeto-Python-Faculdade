```python
import basedosdados as bd
import pandas as pd
```

```python
df = bd.read_table(dataset_id='br_poder360_pesquisas',
table_id='microdados',
billing_project_id="pedro-projetofaculdade-python")
```

```python
pesquisa_2022 = df[(df.instituto == 'Datafolha') &
                  (df.ano == 2022) &
                  (df.cargo == 'presidente') &
                  (df.sigla_uf.isnull()) &
                  (df.tipo == 'estimulada') &
                  ((df.nome_candidato == 'Bolsonaro') | (df.nome_candidato == 'Lula')) &
                   (df.descricao_cenario.str.contains('cenário 1') == True)
                  ].sort_values(by='data')
```
