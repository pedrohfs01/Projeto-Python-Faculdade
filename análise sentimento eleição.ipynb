# Obtendo os dados 

```python
import snscrape.modules.twitter as dados
import pandas as pd
import datetime
import nltk
```

```python
from textblob import TextBlob
from unidecode import unidecode
```

```python
lista_bolsonaro = []
lista_lula = []

sentimento_lula = []
sentimento_bolsonaro = []
```

```python
contador = 1
listaConteudo = []
while contador < 30:
    data_inicial = '2022-10-'+str(contador)
    contador += 1
    data_final = '2022-10-'+str(contador)
    numPos = 0
    numNeg = 0
    for i,tweet in enumerate(dados.TwitterSearchScraper(f'{"Bolsonaro"} + since:{data_inicial} until:{data_final}').get_items()):
        if i>500:
            break
        listaConteudo.append(tweet.content)
        
    df_bolsonaro = pd.DataFrame(listaConteudo, columns=['Text'])
    
    textPT = df_bolsonaro['Text'].apply(unidecode)
    
    for x in textPT:
        sentimento = TextBlob(x) #obtendo o sentimento
        sentimento_bolsonaro.append(sentimento.polarity) #adicionando a uma lista
        if sentimento.polarity > 0: 
            numPos += 1 
        elif sentimento.polarity < 0: 
            numNeg += 1
            
    total = len(sentimento_bolsonaro)
    mediaPos = (total * 100)/numPos
    mediaNeg = (total * 100)/numNeg
    lista_bolsonaro.append([data_inicial, mediaPos, mediaNeg])
```

```python
contador = 1
listaConteudo = []
while contador < 30:
    data_inicial = '2022-10-'+str(contador)
    contador += 1
    data_final = '2022-10-'+str(contador)
    numPos = 0
    numNeg = 0
    for i,tweet in enumerate(dados.TwitterSearchScraper(f'{"Lula"} + since:{data_inicial} until:{data_final}').get_items()):
        if i>500:
            break   
        listaConteudo.append(tweet.content)
        
    df_lula = pd.DataFrame(listaConteudo, columns=['Text'])
    
    textPT = df_lula['Text'].apply(unidecode)
    
    for x in textPT:
        sentimento = TextBlob(x) #obtendo o sentimento
        sentimento_lula.append(sentimento.polarity) #adicionando a uma lista
        if sentimento.polarity > 0: 
            numPos += 1 
        elif sentimento.polarity < 0: 
            numNeg += 1
            
    total = len(sentimento_lula)
    mediaPos = (total * 100)/numPos
    mediaNeg = (total * 100)/numNeg
    lista_lula.append([data_inicial, mediaPos, mediaNeg])
```

```python
df_lula = pd.DataFrame(lista_lula, columns=['Data', 'Média Positiva', 'Média Negativa'])
df_bolsonaro = pd.DataFrame(lista_bolsonaro, columns=['Data', 'Média Positiva', 'Média Negativa'])
```
